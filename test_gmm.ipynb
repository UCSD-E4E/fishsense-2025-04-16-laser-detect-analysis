```python
from pathlib import Path

import cv2 as cv
import matplotlib.pyplot as plt

from fishsense_gmm_laser_detector.gmm_laser_detector import GMMLaserDetector
```

```python
image = Path('P7170081.png')
```

```python
img = cv.imread(image)
```

```python
# denoising???
img_bgr_denoised = cv.fastNlMeansDenoisingColored(img, None, 6, 6, 4, 10)
img_bgr = cv.bilateralFilter(img_bgr_denoised, 10, 75, 75)
```

```python
detector = GMMLaserDetector.initialize_from_json('model.json')
```

```python
x, y = detector.find_laser(img_bgr)
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)

ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
ax.scatter([x], [y], marker='x', s=50)
fig.show()
```
